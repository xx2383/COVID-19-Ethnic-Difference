## Wealth
### Data is from (https://www.federalreserve.gov/econres/scf/dataviz/scf/table/#series:Net_Worth;demographic:racecl4;population:all;units:median;range:1989,2019)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Data of wealth
df = pd.read_csv("C:/Users/yang/Desktop/ucdavis/study/Spring 2021/STA 160/COVID 19 DATASET/NETWORTH.csv",encoding='latin1')
df = df[["year", "Category", "Net_Worth"]]
df = df[df.year >= 2010]
df = df[df.Category != "Other"]
df
```

```python
# Plot wealth disparity across ethnic groups
white = df[df.Category == "White, non-Hispanic"]
black = df[df.Category == "Black, non-Hispanic"]
hispanic = df[df.Category == "Hispanic"]
X = np.arange(4)
fig = plt.subplots(figsize =(12, 8))
plt.bar(X + 0.00, white.Net_Worth, color ='r', width = 0.25, label ='White')
plt.bar(X + 0.25, black.Net_Worth, color ='g', width = 0.25, label ='Black')
plt.bar(X + 0.50, hispanic.Net_Worth, color ='b', width = 0.25, label ='Hispanic')
plt.xticks(X + 0.25, ["2010", "2013", "2016", "2019"])
plt.legend()
plt.title("Median Wealth by Races in US$ Thousands from 2010 to 2019", fontweight ='bold', fontsize = 15)
plt.xlabel("Year", fontweight ='bold', fontsize = 15)
plt.ylabel("US $Thousands", fontweight ='bold', fontsize = 15)
```

## Education Level in 2019

```python
# Make the dataframe of education level data
edu = {"race": ["White alone", "White alone", "White alone", "White alone", "White alone",
             "Non-Hispanic White alone", "Non-Hispanic White alone", "Non-Hispanic White alone", 
                "Non-Hispanic White alone", "Non-Hispanic White alone", 
             "Black alone", "Black alone", "Black alone", "Black alone", "Black alone", 
             "Asian alone", "Asian alone", "Asian alone", "Asian alone", "Asian alone", 
             "Hispanic", "Hispanic", "Hispanic", "Hispanic", "Hispanic"],
        "level": ["High school or more", "College or more", "Associate's degree or more", "Bachelor's degree or more", 
                  "Advanced degree", "High school or more", "College or more", "Associate's degree or more", 
                  "Bachelor's degree or more", "Advanced degree", "High school or more", "College or more", 
                  "Associate's degree or more", "Bachelor's degree or more", "Advanced degree", "High school or more", 
                  "College or more", "Associate's degree or more", "Bachelor's degree or more", "Advanced degree", 
                  "High school or more", "College or more", "Associate's degree or more", "Bachelor's degree or more", 
                  "Advanced degree"],
        "percentage": [0.8978, 0.6151, 0.4368, 0.3363, 0.1201, 
                   0.9376, 0.6629, 0.4817, 0.3758, 0.1368, 
                   0.8723, 0.5399, 0.3348, 0.2374, 0.0829, 
                   0.9106, 0.7360, 0.6133, 0.5470, 0.2257, 
                   0.7300, 0.4139, 0.2461, 0.1676, 0.0487]
      }
edu = pd.DataFrame(edu, columns = ["race", "level", "percentage"])
edu
```

```python
# Plot education level disparity across ethnic groups
plt.figure(figsize = (12,8))
sns.lineplot(x = "level", y = "percentage", hue="race", style = "race", dashes = False, markers = ["o", "<", ">", "^", "v"], 
             data = edu)
plt.title("Education level by races", fontweight ='bold', fontsize = 15)
plt.xlabel("Education levels", fontweight ='bold', fontsize = 15)
plt.ylabel("Percentage", fontweight ='bold', fontsize = 15)
```
