## Year household Income by race from 2015 to 2019

### Data is from https://www.census.gov/data/tables/time-series/demo/income-poverty/cps-pinc/pinc-01.html


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read dataset
df = pd.read_excel("/Users/apple/Desktop/income2019 1.xlsx")
df
```

```python
# Build dataframe
huku = {"race": ["WHITE ALONE","WHITE ALONE","WHITE ALONE","WHITE ALONE","WHITE ALONE","BLACK ALONE OR IN COMBINATION", "BLACK ALONE OR IN COMBINATION", 
                 "BLACK ALONE OR IN COMBINATION", "BLACK ALONE OR IN COMBINATION", "BLACK ALONE OR IN COMBINATION", 
                "BLACK ALONE", "BLACK ALONE", "BLACK ALONE","BLACK ALONE","BLACK ALONE",
             "ASIAN ALONE", "ASIAN ALONE","ASIAN ALONE", "ASIAN ALONE", "ASIAN ALONE"],
        "year": ["2015","2016","2017","2018", "2019","2015","2016","2017","2018", "2019",
                 "2015","2016","2017","2018", "2019",
                 "2015","2016","2017","2018", "2019" ],
        "median income": [64864,65901,67617,68156,72204,
                          40155,42684,41705,42447,46073,
                          39817,42071,41055,42110,45438,
                          83270,86754,84887,88774,98174]
      }
huku = pd.DataFrame (huku, columns = ["race", "year", "median income"])
huku
```

```python
# Plot median income race disparity
plt.figure(figsize = (12,8))
sns.lineplot(x = "year", y = "median income", hue="race", style = "race", dashes = False, 
             data = huku)
plt.title("Year household Income by race from 2015 to 2019", fontweight ='bold', fontsize = 14)
plt.xlabel("Year", fontweight ='bold', fontsize = 15)
plt.ylabel("Median Income", fontweight ='bold', fontsize = 15)
```

Blacks have lower incomes overall and lower growth in income levels for blacks from 2015 to 2019
